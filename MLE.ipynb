```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# function takes in number of observations a,b and c
# and returns the max likelihood, thetas for the max likelihood all likelihood values, the lists of theta values tried
def likelihood(a,b,c):
    #Create the theta space between 0 and 1 with length of a+b+c
    x = np.linspace(0.01, 1, a+b+c)
    theta1s, theta2s, l = [], [], []
    thetas = [0, 0]
    _max = -1000
    for theta1 in x:
        for theta2 in x:
            if theta1 + theta2 > 0.9999:
                break
            theta1s += [theta1]
            theta2s += [theta2]
            #Likelihood formula: a^theta1.b^theta2.c^(1-theta1-theta2)
            l += [(a ** theta1) * (b ** theta2) * (c ** (1 - theta1 - theta2))]
            if l[-1] > _max:
                _max = l[-1]
                thetas = [theta1, theta2]
    return [_max, thetas, l, theta1s, theta2s]
```

```python
# function takes in number of observations a,b and c
# and returns the min log-likelihood, thetas for the min log-likelihood all log-likelihood values, the lists of theta values tried
def loglikelihood(a,b,c):
    #Create the theta space between 0 and 1 with length of a+b+c
    x = np.linspace(0.01, 1, a+b+c)
    theta1s, theta2s, logl = [], [], []
    thetas = [0, 0]
    _min = 0
    for theta1 in x:
        for theta2 in x:
            if theta1 + theta2 > 0.9999:
                break
            theta1s += [theta1]
            theta2s += [theta2]
            #Log-likelihood formula: a.log(theta1) + b.log(theta2) + c.log(1-theta1-theta2)
            logl += [a * np.log(theta1) + b * np.log(theta2) + c * np.log(1 - theta1 - theta2)]
            if logl[-1] < _min:
                _min = logl[-1]
                thetas = [theta1, theta2]
    return [_min, thetas, logl, theta1s, theta2s]
```

```python
#graphing function
#params:
#theta1s: list of theta_1 values used
#theta2s: list of theta_2 values used
#v: list of likelihood values
def graph(theta1s, theta2s, l, title = "graph" ):
    fig = plt.figure(figsize =(10,6))
    ax = plt.axes(projection ="3d")
    ax.scatter3D(theta1s, theta2s,  l, c = [theta1s[k] + theta2s[k] for k in range(len(theta1s))] )
    ax.set_xlabel("theta 1")
    ax.set_ylabel("theta 2")
    if l[0] > 0: ax.set_zlabel("likelihood")
    else: ax.set_zlabel("log-likelihood")
    plt.title(title)
    plt.savefig("{}.png".format(title))
    plt.show()
```

```python
#Question 2
q2 = likelihood(6, 3, 1)
print("Question 2 Likelihood Estimation")
print("Maximum Likelihood:", q2[0])
print("Theta 1:", q2[1][0])
print("Theta 2:", q2[1][1])
graph(q2[3], q2[4], q2[2], "Question 2 Likelihood")
```

```python
#Question 2
q2log = loglikelihood(6, 3, 1)
print("Question 2 Log-Likelihood Estimation")
print("Minimum Log-likelihood:", q2log[0])
print("Theta 1:", q2log[1][0])
print("Theta 2:", q2log[1][1])
graph(q2log[3], q2log[4], q2log[2], "Question 2 Log-likelihood")
```
